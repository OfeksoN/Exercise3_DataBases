### **Correlation Heatmap**
This code generates a heatmap to visualize the correlation between numeric columns in the DataFrame:

* computes the correlation matrix for all numeric columns.
* creates a heatmap.
* renders the plot.

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

### **Scatter Plot with Multiple Dimensions**
This code creates a scatter plot to visualize the relationship between productivity level and daily gaming hours, while incorporating additional dimensions:

plots points with:

* x = horizontal axis.
* y = vertical axis.
* colors points based on performance impact categories.
* adjusts point size according to the gender column.
* sets transparency for better visibility of overlapping points.
* displays the plot.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Productivity_Level', y='Daily_Gaming_Hours', hue='Performance_Impact', data=df, alpha=0.8, size='Occupation')
plt.show()
```

### **Bar Plot of Stress Level by Gender**
This code creates a grouped bar plot to visualize the distribution of stress levels across genders:

* plots the count of each stress level category, with separate bars for each gender.
* add descriptive labels and a title for clarity.
* displays the plot.

```python
sns.countplot(x='Stress_Level', hue='Gender', data=df)
plt.title('Stress_Level by Gender')
plt.xlabel('Stress_Level')
plt.ylabel('Count')
plt.show()
```

### **Game Type Distribution**
This code creates a bar plot to show the distribution of different game types in the dataset:

* sets the figure size for better readability.
* plots the count of each category in the Game_Type column.
* adds a descriptive title.
* rotates the x-axis labels by 45 degrees to prevent overlap and improve clarity.
* displays the plot.

The visualization helps identify which game types are most common in the dataset.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Game_Type')
plt.title('Game Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

### **Performance Impact Distribution (Pie Chart)**
This code creates a pie chart to visualize the proportion of each category in the Performance_Impact column:    

calculates the frequency of each category.  
draws the pie chart with:

* sizes of the slices.
* category names as labels.
* displays percentages with one decimal place.
* rotates the chart so the first slice starts at the top.
* adds a descriptive title.
* renders the chart.

This visualization helps understand the relative distribution of performance impact categories in the dataset.

```python
impact_counts = df["Performance_Impact"].value_counts()
plt.figure()
plt.pie(
    impact_counts,
    labels=impact_counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Performance Impact Distribution")
plt.show()
```

### **Performance Impact by Gender**

This code creates a bar plot to visualize the distribution of performance impacts across genders:

* Creates a bar chart that counts how many rows are in each category of Performance_Impact (x-axis).
* splits each category into sub-bars by Gender, so you can compare genders within each impact level.
* Adds a title and axis labels to make the chart readable.
* Displays the chart inline in the Colab notebook output cell.

```python
sns.countplot(x='Performance_Impact', hue='Gender', data=df)

plt.title('Performance Impact by Gender')
plt.xlabel('Performance Impact')
plt.ylabel('Count')
plt.show()
```
